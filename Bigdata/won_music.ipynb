```python
from selenium import webdriver
from selenium.webdriver import ActionChains
import time
import numpy as np
import pandas as pd
from tqdm import trange, tqdm, tqdm_notebook
import re
```

```python
driver = webdriver.Chrome('C:/Temp/chromedriver')
```

```python
URL = 'https://gongu.copyright.or.kr/gongu/wrt/wrtCl/listWrtSound.do?menuNo=200020&pageIndex='
URL_ = "li:nth-child("
URL_title = ") > div:nth-child(1) > div.col-md-8 > span.tit > a"
URL_tag1 = ") > div:nth-child(3) > div.col-md-9 > div > ul > li:nth-child("
URL_tag2 = ") > a"
URL_recommend = ") > div:nth-child(3) > div.col-md-3 > div > a.isep.icon_reco.recommendBtn > span.recNum"
URL_click = ") > div:nth-child(3) > div.col-md-3 > div > span"
URL_download = ") > div:nth-child(3) > div.col-md-3 > div > a:nth-child(3) > span"

title = []
tag1 = []
tag2 = []
tag3 = []
tag4 = []
tag5 = []
tag6 = []
tag7 = []
tag8 = []
tag9 = []
tag10 = []
tag11 = []
tag12 = []
recommend = []
clicks = []
download = []

for i in tqdm(range(1, 655)):
    driver.get(URL + str(i))
    time.sleep(1)
    for j in range(1, 25):
        title.append(driver.find_element_by_css_selector(URL_ + str(j) + URL_title).text)
        recommend.append(driver.find_element_by_css_selector(URL_ + str(j) + URL_recommend).text)
        clicks.append(driver.find_element_by_css_selector(URL_ + str(j) + URL_click).text)
        download.append(driver.find_element_by_css_selector(URL_ + str(j) + URL_download).text)
        try:
            tag1.append(driver.find_element_by_css_selector(URL_ + str(j) + URL_tag1 + '1' + URL_tag2).text)
        except:
            tag1.append(np.NaN)
        try:
            tag2.append(driver.find_element_by_css_selector(URL_ + str(j) + URL_tag1 + '2' + URL_tag2).text)
        except:
            tag2.append(np.NaN)
        try:
            tag3.append(driver.find_element_by_css_selector(URL_ + str(j) + URL_tag1 + '3' + URL_tag2).text)
        except:
            tag3.append(np.NaN)
        try:
            tag4.append(driver.find_element_by_css_selector(URL_ + str(j) + URL_tag1 + '4' + URL_tag2).text)
        except:
            tag4.append(np.NaN)
        try:
            tag5.append(driver.find_element_by_css_selector(URL_ + str(j) + URL_tag1 + '5' + URL_tag2).text)
        except:
            tag5.append(np.NaN)
        try:
            tag6.append(driver.find_element_by_css_selector(URL_ + str(j) + URL_tag1 + '6' + URL_tag2).text)
        except:
            tag6.append(np.NaN)
        try:
            tag7.append(driver.find_element_by_css_selector(URL_ + str(j) + URL_tag1 + '7' + URL_tag2).text)
        except:
            tag7.append(np.NaN)
        try:
            tag8.append(driver.find_element_by_css_selector(URL_ + str(j) + URL_tag1 + '8' + URL_tag2).text)
        except:
            tag8.append(np.NaN)
        try:
            tag9.append(driver.find_element_by_css_selector(URL_ + str(j) + URL_tag1 + '9' + URL_tag2).text)
        except:
            tag9.append(np.NaN)
        try:
            tag10.append(driver.find_element_by_css_selector(URL_ + str(j) + URL_tag1 + '10' + URL_tag2).text)
        except:
            tag10.append(np.NaN)
        try:
            tag11.append(driver.find_element_by_css_selector(URL_ + str(j) + URL_tag1 + '11' + URL_tag2).text)
        except:
            tag11.append(np.NaN)
        try:
            tag12.append(driver.find_element_by_css_selector(URL_ + str(j) + URL_tag1 + '12' + URL_tag2).text)
        except:
            tag12.append(np.NaN)
```

```python
music = pd.DataFrame({'title' : title, 'recommend' : recommend, 'clicks' : clicks, 'download' : download,
                      'tag1' : tag1, 'tag2' : tag2, 'tag3' : tag3, 'tag4' : tag4, 'tag5' : tag5, 'tag6' : tag6,
                      'tag7' : tag7, 'tag8' : tag8, 'tag9' : tag9, 'tag10' : tag10, 'tag11' : tag11, 'tag12' : tag12})
```

```python
music['clicks'] = music['clicks'].apply(lambda x: re.sub('조회수\n', '', x))
music['download'] = music['download'].apply(lambda x: re.sub('다운로드 수\n', '', x))
music
```

```python
music.to_csv('music.csv', encoding = "UTF-8")
```
