```python
import pandas as pd
```

```python
music = pd.read_csv('music.csv').iloc[:, 1:]
music
```

```python
music = music[music['tag1'].isna() == False]
music
```

```python
def viewtag(music, string) :
    return music[(music['tag1'] == string) | (music['tag2'] == string) | (music['tag3'] == string) | (music['tag4'] == string) |
                 (music['tag5'] == string) | (music['tag6'] == string) | (music['tag7'] == string) | (music['tag8'] == string) |
                 (music['tag9'] == string) | (music['tag10'] == string)]
```

```python
def deltag(music, string) :
    return music[(music['tag1'] != string) & (music['tag2'] != string) & (music['tag3'] != string) & (music['tag4'] != string) &
                 (music['tag5'] != string) & (music['tag6'] != string) & (music['tag7'] != string) & (music['tag8'] != string) &
                 (music['tag9'] != string) & (music['tag10'] != string)]
```

```python
music[(music['tag1'] == '차이나') | (music['tag2'] == '차이나') | (music['tag3'] == '차이나') | (music['tag4'] == '차이나')
     | (music['tag5'] == '차이나') | (music['tag6'] == '차이나') | (music['tag7'] == '차이나') | (music['tag8'] == '차이나') | 
     (music['tag9'] == '차이나') | (music['tag10'] == '차이나')]
```

```python
viewtag(music, '느리게')
```

```python
# 차이나 보이스(VOICE) 1950년대음반 국악(KOREAN TRADITIONAL) 보이스 목소리 효과음 기계음 소음 유성기음반 컬럼비아레코드
```

```python
music1 = deltag(music, '차이나')
music1 = deltag(music1, '보이스(VOICE)')
music1 = deltag(music1, '1950년대음반')
music1 = deltag(music1, '국악(KOREAN TRADITIONAL)')
music1 = deltag(music1, '보이스')
music1 = deltag(music1, '목소리')
music1 = deltag(music1, '효과음')
music1 = deltag(music1, '기계음')
music1 = deltag(music1, '소음')
music1 = deltag(music1, '유성기음반')
music1 = deltag(music1, '컬럼비아레코드')
music1 = deltag(music1, '프랑스')
music1 = deltag(music1, '오스트리아')
music1 = deltag(music1, '헝가리')
music1 = deltag(music1, '이지리스닝')
music1 = deltag(music1, '코믹&키즈')
music1 = deltag(music1, '바로크')
```

```python
viewtag(viewtag(music1, "밝은"), "느리게")
```
